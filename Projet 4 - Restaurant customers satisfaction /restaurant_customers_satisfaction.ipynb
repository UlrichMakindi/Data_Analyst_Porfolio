```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
### Import DataBase from local import

df = pd.read_csv('/Users/um/Desktop/Data_Analyst_Portfolio/Projet 4 - Restaurant customers satisfaction /restaurant_customer_satisfaction.csv')
```

```python
df.head()
```

```python
### Checking for missing values and NaN values for each columns 
print(df.isnull().sum())
```

```python
### Drop missing values
df = df.dropna(axis =1)
```

```python
### Explore data
df.describe().T
```

```python
### Check remaining columns after removing missing values
df.isnull().sum()
```

```python
# Age distrubution of customers
plt.figure(figsize=(10,6))
plt.title('Age distribution of customers')
plt.xlabel('Age')
plt.ylabel('Count')
sns.set_style("whitegrid")
sns.histplot(df['Age'], bins = 10, kde = True, alpha = 0.5)
sns.set_style("whitegrid")
plt.show()
```

```python
### Adjusting figure size
plt.figure(figsize=(10, 6))

# Creating the improved barplot
ax = sns.barplot(x="Gender", y="Age", data=df, hue="PreferredCuisine", ci=None, palette="Set2")

### Adding title and axis labels
plt.title('Average Age by Gender and Preferred Cuisine', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Average Age', fontsize=12)

# Adding numbers on top of each bar
for container in ax.containers:
    ax.bar_label(container, fmt='%.1f', label_type='edge', padding=3)  # '%.1f' formats to one decimal

### Moving the legend outside the plot
plt.legend(title='PreferredCuisine', bbox_to_anchor=(1.05, 1), loc='upper left')

### Displaying the improved plot
plt.tight_layout()
plt.grid(False)
plt.show()
```

```python
### Adjusting figure size for side-by-side plots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

### Barplot for TimeOfVisit
sns.barplot(ax=axes[0], x="Gender", y="Income", data=df, hue="TimeOfVisit", ci=None, palette="Set2")
axes[0].set_title('Income by Gender and Time of Visit', fontsize=16)
axes[0].set_xlabel('Gender', fontsize=12)
axes[0].set_ylabel('Average Income', fontsize=12)
axes[0].legend(title='TimeOfVisit', bbox_to_anchor=(1.05, 1), loc='upper left')

### Adding value labels to bars in TimeOfVisit plot
for container in axes[0].containers:
    axes[0].bar_label(container, fmt='%.1f', label_type='edge', padding=3)

# Barplot for DiningOccasion
sns.barplot(ax=axes[1], x="Gender", y="Income", data=df, hue="DiningOccasion", ci=None, palette="Set2")
axes[1].set_title('Income by Gender and Dining Occasion', fontsize=16)
axes[1].set_xlabel('Gender', fontsize=12)
axes[1].set_ylabel('Average Income', fontsize=12)
axes[1].legend(title='DiningOccasion', bbox_to_anchor=(1.05, 1), loc='upper left')

### Adding value labels to bars in DiningOccasion plot
for container in axes[1].containers:
    axes[1].bar_label(container, fmt='%.1f', label_type='edge', padding=3)

### Adjusting layout
plt.tight_layout()
plt.show()
```

```python
### WaitTime comparison by customer satisfaction.

plt.figure(figsize=(10,6))
sns.boxplot(x='HighSatisfaction', y='WaitTime', data=df, palette='Set2')
plt.title('WaitTime Comparison by Customer Satisfaction', fontsize=16)
plt.xlabel('High Satisfaction (1 = Yes, 0 = No)', fontsize=12)
plt.ylabel('WaitTime (minutes)', fontsize=12)
plt.grid(False)
plt.show()
```

```python
# Correlation matrix
corr_matrix = df[['WaitTime', 'HighSatisfaction', 'Age', 'Income', 'AverageSpend', 'GroupSize', 'OnlineReservation', 'DeliveryOrder', 'LoyaltyProgramMember',
                  'ServiceRating', 'FoodRating', 'AmbianceRating', 'HighSatisfaction']].corr()

# Show the correlation matrix
plt.figure(figsize=(18,10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation between WaitTime and HighSatisfaction', fontsize=16)
plt.show()
```

```python
# Adjusting figure size
plt.figure(figsize=(15, 12))

# Plot 1: WaitTime vs FoodRating
plt.subplot(3, 3, 1)
sns.scatterplot(x='WaitTime', y='FoodRating', data=df)
plt.title('WaitTime vs FoodRating (corr = 0.37)')
plt.xlabel('WaitTime')
plt.ylabel('FoodRating')

# Plot 2: HighSatisfaction vs Income
plt.subplot(3, 3, 2)
sns.scatterplot(x='HighSatisfaction', y='Income', data=df)
plt.title('HighSatisfaction vs Income (corr = 0.82)')
plt.xlabel('HighSatisfaction')
plt.ylabel('Income')

# Plot 3: HighSatisfaction vs AverageSpend
plt.subplot(3, 3, 3)
sns.scatterplot(x='HighSatisfaction', y='AverageSpend', data=df)
plt.title('HighSatisfaction vs AverageSpend (corr = 0.53)')
plt.xlabel('HighSatisfaction')
plt.ylabel('AverageSpend')

# Plot 4: HighSatisfaction vs AmbianceRating
plt.subplot(3, 3, 4)
sns.scatterplot(x='HighSatisfaction', y='AmbianceRating', data=df)
plt.title('HighSatisfaction vs AmbianceRating (corr = 0.78)')
plt.xlabel('HighSatisfaction')
plt.ylabel('AmbianceRating')

# Plot 5: Income vs HighSatisfaction
plt.subplot(3, 3, 5)
sns.scatterplot(x='Income', y='HighSatisfaction', data=df)
plt.title('Income vs HighSatisfaction (corr = 0.82)')
plt.xlabel('Income')
plt.ylabel('HighSatisfaction')

# Plot 6: AverageSpend vs HighSatisfaction
plt.subplot(3, 3, 6)
sns.scatterplot(x='AverageSpend', y='HighSatisfaction', data=df)
plt.title('AverageSpend vs HighSatisfaction (corr = 0.53)')
plt.xlabel('AverageSpend')
plt.ylabel('HighSatisfaction')

# Plot 7: AverageSpend vs AmbianceRating
plt.subplot(3, 3, 7)
sns.scatterplot(x='AverageSpend', y='AmbianceRating', data=df)
plt.title('AverageSpend vs AmbianceRating (corr = 0.95)')
plt.xlabel('AverageSpend')
plt.ylabel('AmbianceRating')

# Plot 8: GroupSize vs Income
plt.subplot(3, 3, 8)
sns.scatterplot(x='GroupSize', y='Income', data=df)
plt.title('GroupSize vs Income (corr = 0.63)')
plt.xlabel('GroupSize')
plt.ylabel('Income')

# Plot 9: AmbianceRating vs HighSatisfaction
plt.subplot(3, 3, 9)
sns.scatterplot(x='AmbianceRating', y='HighSatisfaction', data=df)
plt.title('AmbianceRating vs HighSatisfaction (corr = 0.78)')
plt.xlabel('AmbianceRating')
plt.ylabel('HighSatisfaction')

# Adjusting layout to avoid overlap
plt.tight_layout()
plt.show()
```
